```python
#!pip install visualkeras
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, Dense, Dropout, Flatten, MaxPooling2D, Conv3D
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import ModelCheckpoint
import visualkeras
```

```python
train_data=ImageDataGenerator(rescale = 1./255)
test_data= ImageDataGenerator(rescale = 1./255)
```

```python
batch_size=50
train_data=train_data.flow_from_directory(
'Dataset/train/',
target_size = (32,32),
batch_size=batch_size,
class_mode='categorical')

test_data=test_data.flow_from_directory(
'Dataset/test/',
target_size = (32,32),
batch_size=batch_size,
class_mode='categorical')
```

```python
train_data.image_shape
```

```python
model = keras.Sequential()
model.add(Conv2D(32, kernel_size=(3, 3), activation='relu', input_shape=(32, 32, 3)))
model.add(Conv2D(64, (3, 3), activation='relu'))
model.add(MaxPooling2D(pool_size=(2, 2)))
model.add(Dropout(0.5))
model.add(Conv2D(128, (3, 3), activation='relu'))
model.add(MaxPooling2D(pool_size=(2, 2)))
model.add(Conv2D(128, (3, 3), activation='relu'))
model.add(MaxPooling2D(pool_size=(2, 2)))
model.add(Dropout(0.5))
model.add(Flatten())
model.add(Dense(1024, activation='relu',kernel_regularizer='l2'))
model.add(Dropout(0.5))
model.add(Dense(7, activation='softmax',kernel_regularizer='l2'))
```

```python
model.summary()
```

```python
model.compile(optimizer=Adam(learning_rate=0.0001), loss='categorical_crossentropy', metrics=['acc'])
```

```python
visualkeras.layered_view(model, to_file='output.png').show()
```

```python
checkpoint = ModelCheckpoint('model2-{epoch:03d}.model',monitor='val_loss',
                             verbose=0,save_best_only=True,mode='auto')
```

```python
history=model.fit(train_data,epochs=300,validation_data=test_data,steps_per_epoch=399)
```
